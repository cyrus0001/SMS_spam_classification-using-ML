```python
import pandas as pd 
```

```python
messages = pd.read_csv('SMSSpamCollection.txt', sep = '\t',names = ["Label","Message"])
```

```python
messages
```

```python
messages.shape
```

```python
messages['Message'].loc[100]
```

```python
import re 
import nltk 
```

```python
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
corpus = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['Message'][i]) 
    review = review.lower()
    review = review.split()
    review = [word for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500,binary = True,ngram_range = (2,2))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
y = pd.get_dummies(messages['Label'])
y = y.iloc[: , 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.15 , random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detection_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

score = accuracy_score(y_test, y_pred)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 2500,binary = True , ngram_range = (2,2))
X_1 = tfidf.fit_transform(corpus).toarray()
```

```python
X_1[1]
```

```python
X_1_train,X_1_test,y_1_train,y_1_test = train_test_split(X_1,y,test_size = 0.15,random_state = 0 )
```

```python
spam_detection_model_tfidf = MultinomialNB().fit(X_1_train,y_1_train)
```

```python
y_pred_tfidf = spam_detection_model_tfidf.predict(X_1_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score_tfidf = accuracy_score(y_1_test,y_pred_tfidf)
print(score_tfidf)
```

```python
print(classification_report(y_pred_tfidf,y_1_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(accuracy_score(y_pred,y_test))
print(classification_report(y_test,y_pred))
```
